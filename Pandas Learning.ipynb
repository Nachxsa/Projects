```python
import pandas as pd
import numpy as np
```

#### Adding / Removing Columns

```python
'''Remove Row or Column'''
# coffee.drop(columns=['Coffee Type'], inplace=True)
# coffee.drop(0, inplace=True)
# coffee.head()

'''Copying a dataframe'''
coffe_new = coffee.copy()

'''Adding Columns'''
coffe_new['price'] = 4.99

'''Changing dates'''
# bios_new['datetime'] = pd.todatetime(bios_new['born_date'], format="")

'''Converting dataframs to csv'''
coffe_new.to_csv('coffee_new.csv', index=False) #The index is for removing the first column which has the index for the rows

'''Applying functions to columns'''
coffe_new['price'] = coffe_new['price'].apply(lambda x: x * 2)
```

#### Merging and Concatenating Data

```python
bios = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv')
nocs = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv')
```

```python
'''Merge Data'''
bios_new = pd.merge(bios, # Dataframe a unir 1
                    nocs, # Dataframe a unir 2
                    left_on='born_country', # La columna de bios a unir
                    right_on='NOC', # La columna de nocs a unir. Si las columnas tuviesene l msimo nombre, se usa 'on' solo
                    how='inner',
                    suffixes=["1", '2']) #En caso de que las columnas tengan el msimo nombre, se le agrega eso al final
# Parámetro how:
# inner: Solo las filas con coincidencia en ambas tablas. (Elimina lo que no coincida).
# left:Todas las filas del primer DataFrame (left), con NaN donde no haya coincidencia en el segundo.
# right:Todas las filas del segundo DataFrame (right), con NaN donde no haya coincidencia en el primero.
# outer:Todas las filas de ambos DataFrames, con NaN donde no haya coincidencia.

bios_new.head(2)
```

#### Handlig null values

```python
coffee.loc[[2,3], 'Units Sold'] = np.nan

'''fill na'''
# coffee.fillna(0)
# coffee.fillna(coffee['Units Sold'].mean())
coffee['Units Sold'] = coffee['Units Sold'].interpolate() #looks for patterns
coffee.head()

'''Drop Na Values'''
coffee.dropna() # Drop all Na Values
coffee.dropna(subset=['Units Sold'], inplace=True) # Drop na only for Na values in Units Sold
```

#### Aggregating Data

```python
'''Count How manny Values there are for an specifc Values or Caracteristics'''
bios[bios['born_country']=='USA']['born_region'].value_counts()

'''Group By'''
coffee.groupby(['Coffee Type'])['Units Sold'].sum()

'''Group by and trigger functions for each row'''
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum'})
```
